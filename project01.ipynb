```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics as stac
data=pd.read_csv('housingtest.csv')
latitude=data.latitude.to_numpy()
rooms=data.rooms.to_numpy()
plt.scatter(latitude,rooms)
print("mean of latitude is",stac.mean(latitude))
print("mean of rooms is",stac.mean(rooms))
print("mode of latitude is",stac.mode(latitude))
#print("mode of rooms is",stac.mode(rooms))
print("variance of latitude is",stac.pvariance(latitude))
print("variance of rooms is",stac.pvariance(rooms))
print("standard deviation of latitude is",stac.pstdev(latitude))
print("standard deviation of rooms is",stac.pstdev(rooms))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics as stac
from scipy import stats
data=pd.read_csv('housingtest.csv')
population=data.population.to_numpy()
households=data.households.to_numpy()
slope, intercept, r, p, std_err = stats.linregress(population,households)
def myfunc(population):
  return slope * population + intercept

mymodel = list(map(myfunc, population))

plt.scatter(population,households)
plt.plot(population, mymodel)
plt.show()

print("mean of population is",stac.mean(population))
print("mean of households is",stac.mean(households))
print("mode of population is",stac.mode(population))
#print("mode of households is",stac.mode(households))
print("variance of population is",stac.pvariance(population))
print("variance of households is",stac.pvariance(households))
print("standard deviation of population is",stac.pstdev(population))
print("standard deviation of households is",stac.pstdev(households))
```
